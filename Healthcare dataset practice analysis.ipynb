# LET'S DIVE INTO DATA ANALYSIS PART

# 1. SCATTER PLOT

```python
df_sorted = df.sort_values(by = 'Billing Amount',ascending=False)
```

```python
df_top_100 = df_sorted.head(100)
```

```python
df['Age'] = pd.to_numeric(df['Age'],errors='coerce') 
```

```python
#create scatter plot
plt.figure(figsize=(10,6))

plt.scatter(df_top_100['Age'],df_top_100['Billing Amount'])

plt.title("Age vs Billing Amount (Top-100)")
plt.xlabel("Age")
plt.ylabel("Billing Amount")
plt.show()
```

# 2. PAIR-PLOT

```python
#create a pairplot
plt.figure(figsize=(12,8))
sns.pairplot(df_top_100, vars=['Age', 'Billing Amount'], hue = 'Gender')

plt.suptitle("Age,Billing Amount by Gender",y=1.02)
plt.show()
```

# 3 HISTOGRAM

```python
#create histogram
plt.figure(figsize=(10,6))

sns.histplot(df_top_100['Billing Amount'], bins=10, kde=True , color='skyblue')

plt.title('Histogram')
plt.xlabel('Age')
plt.ylabel("Billing Amount")
plt.show()
```
